```python
from math import floor

text = open("input.txt").read().split("\n")
direction_map = ['R', 'D', 'L', 'U']
text = [x.split(" ")[2] for x in text]
text = [(direction_map[int(x[7])], int(x[2:7], 16)) for x in text]
text
```

```python
directions = {
    'R': (0, 1),
    'D': (1, 0),
    'L': (0, -1),
    'U': (-1, 0)
}

def tsum(one, two, factor):
    return (one[0] + two[0] * factor, one[1] + two[1] * factor)

def area(points):
    total_area = 0
    for i, point in enumerate(points):
        next = (i + 1) % len(points)
        total_area += ((point[0] * points[next][1]) - (point[1] * points[next][0]))
    return total_area / 2

loop = [(0, 0)]

length = 0
for order in text:
    direction, distance = order
    direction = directions[direction]
    current = loop[-1]
    length += distance
    loop.append(tsum(current, direction, distance))
    

A = area(loop[::-1])
b = length / 2
i = A - b + 1
print(floor(i + length))
```
