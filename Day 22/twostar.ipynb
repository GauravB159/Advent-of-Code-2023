```python
import numpy as np
import matplotlib.pyplot as plt
text = open("input.txt").read().split("\n")
text = [[[int(z) for z in y.split(",")] for y in x.split("~")] for x in text]

sort_by_height = sorted([(min(x[0][2], x[1][2]), i) for i, x in enumerate(text)])
print(sort_by_height)
min_at_xy = {}
grid = np.zeros((400, 400, 400))
supports = {i: set() for i, x in enumerate(text)}
inverse_supports = {i: set() for i, x in enumerate(text)}
for point in sort_by_height:
    left, right = text[point[1]]
    difference = [x[1] - x[0] for x in list(zip(left, right))]
    check_difference = [1 if x > 0 else 0 for x in difference]
    orientation = [i for i,x in enumerate(difference) if x]
    orientation = orientation[0] if len(orientation) > 0 else 0
    points = [(left[0] + x * check_difference[0],left[1] + x * check_difference[1],left[2] + x * check_difference[2]) for x in range(difference[orientation] + 1)]
    for k in range(1, point[0] + 1):
        invalid = False
        for i in range(difference[orientation] + 1):
            if grid[left[0] + i * check_difference[0]][left[1] + i * check_difference[1]][left[2] - k + i * check_difference[2]] != 0:
                supporter = int(grid[left[0] + i * check_difference[0]][left[1] + i * check_difference[1]][left[2] - k + i * check_difference[2]])
                supported = point[1]
                supports[supporter - 1].add(supported)
                inverse_supports[supported].add(supporter - 1)
                invalid = True
        if invalid: 
            break
        points = []
        for i in range(difference[orientation] + 1):
            points.append((left[0] + i * check_difference[0],left[1] + i * check_difference[1],left[2] - k + i * check_difference[2]))
    for pnt in points:
        grid[pnt[0]][pnt[1]][pnt[2]] = point[1] + 1
        
print(supports)
print()
print(inverse_supports)
removable = []


def bfs(point):
    queue = [point]
    moved = {}
    while len(queue):
        point = queue.pop(0)
        moved[point] = 1
        for pt in supports[point]:
            spt_count = 0
            for support in inverse_supports[pt]:
                if moved.get(support, 0):
                    continue
                spt_count += 1
            if spt_count == 0:
                queue.append(pt)
    return len(moved) - 1

result = 0
for point in supports:
    result += bfs(point)

print(result)
```
