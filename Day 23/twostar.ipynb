```python
import math

text = open("input.txt").read().split("\n")
text = [list(x) for x in text]
grid = {}
visited = {}
parents = {}
actual_distances = {}
total_count = 0
for i, row in enumerate(text):
    for j, col in enumerate(row):
        if i == 0:
            if col == '.':
                start  = (i, j)
        elif i == len(text) - 1:
            if col == '.':
                end = (i, j)
        total_count += 1
        grid[(i, j)] = col
        visited[(i, j)] = 0
        actual_distances[(i, j)] = math.inf

directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]

slopes = {
    'v': (1, 0),
    '^': (-1, 0),
    '>': (0, 1),
    '<': (0, -1)
}

queue = [(start, 0, (-1, -1), -1)]

maximum = 0

visited = {}
intersections = {}
while queue:
    point, distance, odirection, split = queue.pop(0)
    if not grid.get(point, False):
        continue
    if grid[point] == '#':
        continue
    if (point, split) in visited:
        continue
    if point == end:
        if split not in intersections:
            intersections[split] = []
        intersections[split].append((end, distance))
    visited[(point, split)] = True
    splits = [grid.get((point[0] + direction[0], point[1] + direction[1])) for direction in directions if ((direction[0] + odirection[0] != 0) or (direction[1] + odirection[1] != 0))]
    splits = [x for x in splits if x in slopes or x == '.']
    if len(splits) > 1 and point != split:
        if split not in intersections:
            intersections[split] = []
        intersections[split].append((point, distance))
        split = point
        distance = 0
    for direction in directions:
        queue.append(((point[0] + direction[0], point[1] + direction[1]), distance + 1, direction, split))

intersections
```
