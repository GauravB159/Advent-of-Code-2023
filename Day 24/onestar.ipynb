```python
text = open("input.txt").read().split("\n")
text = [[[int(z) for z in y.split(", ")] for y in x.split(" @ ")] for x in text]
crossed = []
count = 0
for i, one in enumerate(text):
    for j, two in enumerate(text[i + 1:]):
        coord_one, delta_one = one
        x1, y1, z1 = coord_one
        dx1, dy1, dz1 = delta_one
        coord_two, delta_two = two
        x2, y2, z2 = coord_two
        dx2, dy2, dz2 = delta_two
        m_one = dy1 / dx1
        m_two = dy2 / dx2
        line_one = (m_one, y1 - m_one * x1)
        line_two = (m_two, y2 - m_two * x2)

        div = dx1 * dy2 - dy1 * dx2
        if div == 0:
            continue
        
        d1 = x1 * (y1 + dy1) - y1 * (x1 + dx1)
        d2 = x2 * (y2 + dy2) - y2 * (x2 + dx2)
        x = -(d1 * dx2 - dx1 * d2) / div
        y = -(d1 * dy2 - dy1 * d2) / div
        if x >= 200000000000000 and x <= 400000000000000 and y >= 200000000000000 and y <= 400000000000000:
            if dx1 > 0 and x < x1:
                continue
            if dx2 > 0 and x < x2:
                continue
            if dx1 < 0 and x > x1:
                continue
            if dx2 < 0 and x > x2:
                continue
            if dy1 > 0 and y < y1:
                continue
            if dy2 > 0 and y < y2:
                continue
            if dy1 < 0 and y > y1:
                continue
            if dy2 < 0 and y > y2:
                continue
            count += 1
            print(i, i + j + 1, x, y)

print(count, crossed)
```
